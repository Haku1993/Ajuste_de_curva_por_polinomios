# Tarea 1, replicar gr√°ficos de las clases

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(12345)
x = np.arange(0, 1, 0.02)
y = np.sin(np.pi*x*2)
z1 = np.arange(0, 1, 0.1)
u = np.random.randn(10)/5
z2 = np.sin(np.pi*z1*2)+u
plt.plot(x,y)
plt.plot(z1,z2,"ro")
plt.show()
```

```python
poly0 = np.poly1d(np.polyfit(z1,z2,9))
xp = np.linspace(0, 0.9, 1000)
plt.plot(x,y)
plt.plot(z1,z2,"ro")
plt.plot(xp,poly0(xp))
plt.show()
```

Los anteriores c√≥digos crean los plots de la funci√≥n seno con sus respectivos puntos que contienen ruido y polynomial curve fitting, procedemos a comparar con polinomios de grado superior.

```python
fig, ax = plt.subplots(2, 2,sharex = True, sharey = True, figsize=(12,10))

poly00 = np.poly1d(np.polyfit(z1,z2,0))
ax[0,0].plot(x,y)
ax[0,0].plot(z1,z2,"ro")
ax[0,0].plot(xp,poly00(xp))
ax[0,0].text(0.8,0.8,"M = 0",fontsize=15)

poly01 = np.poly1d(np.polyfit(z1,z2,1))
ax[0,1].plot(x,y)
ax[0,1].plot(z1,z2,"ro")
ax[0,1].plot(xp,poly01(xp))
ax[0,1].text(0.8,0.8,"M = 1",fontsize=15)

poly10 = np.poly1d(np.polyfit(z1,z2,3))
ax[1,0].plot(x,y)
ax[1,0].plot(z1,z2,"ro")
ax[1,0].plot(xp,poly10(xp))
ax[1,0].text(0.8,0.8,"M = 3",fontsize=15)

poly11 = np.poly1d(np.polyfit(z1,z2,9))
ax[1,1].plot(x,y)
ax[1,1].plot(z1,z2,"ro")
ax[1,1].plot(xp,poly11(xp))
ax[1,1].text(0.8,0.8,"M = 9",fontsize=15)

fig.suptitle("Polynomial Curve Fitting",fontsize=20)
```
